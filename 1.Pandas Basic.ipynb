## Handling missing values - isnull(), dropna() using subset, fillna()

```python
ufo = pd.read_csv('http://bit.ly/uforeports') 
ufo.head()
```

```python
ufo.tail()
```

```python
ufo.isnull().sum()
```

```python
ufo.shape
```

```python
ufo.City.isnull().sum()
```

```python
ufo[ufo.City.isnull()]
```

```python
ufo.dropna(how='any').shape
```

```python
ufo.dropna(how='all').shape
```

```python
# use certains columns (use subset menthod) to dropna
ufo.dropna(subset=['Colors Reported','Shape Reported'], how = 'all').shape 
```

```python
# fillna is also working for a series
ufo['Shape Reported'].fillna(value = 'VARIOUS')
```
