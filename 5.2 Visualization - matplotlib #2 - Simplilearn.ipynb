```python
from matplotlib import pyplot as plt # this is more popular usage
```

```python
x = np.linspace(0, 10, 25) # one useful numpy function to geneate random series
y = x*x + 2
```

```python
fig = plt.figure()
axis = fig.add_axes([0.5, 0.1, 0.8, 0.8])
axis.plot(x, y, 'r')
```

```python
fig = plt.figure(figsize=(16,9), dpi = 80)  ## figsize, dpi para to enlarge the paint
axes1 = fig.add_axes([0.1, 0.1, 0.5, 0.8])
axes2 = fig.add_axes([0.1, 0.6, 0.4, 0.3])
axes1.plot(x, y, 'r')
axes2.plot(y, x, 'g')
```

## line color, transparency, and use fill_between to fill with different color
### try to put all the 'painting' actions into SAME cell, otherwise it may not work to show the diagram

```python
fig, axes = plt.subplots(dpi = 100)

axes.set_title('Plot trial - fill_between, and theme')
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')

axes.plot(x, x+10, 'r', marker ='o', markersize = 5, label ='y = x + 10')
axes.plot(x, x*x-2*x, 'g', marker ='*',  markersize = 5,  label ='y = x*x - 2*x')
axes.plot(x, x*x-10, 'b', marker ='*',  markersize = 5,  label ='y = x*x + 10')

axes.fill_between(x, x+10, x*x-2*x, 
                  where = ((x+10)> (x*x-2*x)),
                  color = 'r',
                  interpolate = True,
                  alpha = 0.5)

axes.fill_between(x, x+10, x*x-2*x, 
                  where = ((x+10)< (x*x-2*x)),
                  color = 'g',
                  interpolate = True,
                  alpha = 0.5)

axes.grid()
axes.legend()
```
